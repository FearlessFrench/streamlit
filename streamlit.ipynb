# <b>Streamlit</b>

#### A faster way to build and share data apps. Turn your data scripts into shareable web apps in minutes.

All in pure Python. No front‑end experience required.

### <b>1. Install Streamlit</b>

```python
# You can install it with pip using this command
!pip install streamlit
```

### <b>2. Create a Streamlit Script</b>

Create a `.py` file (e.g., `app.py`) and add some sample code:
<hr>
<pre><code>import streamlit as st

st.title("Hello Streamlit!")
st.write("This is my first Streamlit app in VS Code.")</code></pre>

### <b>3. Run the App</b>
<ol>
    <li>Right Click on Streamlit Folder and select Open in Integrated Terminal</li>
    <li>Add this line of code "streamlit run app.py" (It will launch a local server (usually at http://localhost:8501) and open it in your browser.)</li>
